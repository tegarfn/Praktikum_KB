<h2><b>Visualisasi Data</b></h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv('tv_shows.csv')
ds
```

<h3><b>Membuat Grafik Batang</b></h3>
<p>
    untuk membut grafik batang, menggunakan fungsi 'plt.bar()', carannya sebagai berikut: 
</p>
<ol>
    <li>memanggil fungsi plt.bar()</li>
    <li>fungsi diisi dengan data pertama untuk line X, data kedua untuk line Y, dan warna grafiknya</li>
    <li>berikan label grafik, label Y, dan label Y</li>    
</ol>
<h4><b>Hasil</b></h4>
<p>
    dari hasil visualisasi data, didapat bahwa film dengan rating 'all' paling banyak tampil di platform film <br>
    rating '13+' mulai muncul di tahun 90-an hingga sekarang<br>
</p>

```python
ds.dropna(inplace=True)

plt.bar(ds['Year'], ds['Age'], color='red')
plt.title('Rating berdasarkan Tahun Rilis', loc='center')
plt.xlabel('Tahun')
plt.ylabel('Rating')
plt.show()
```

<h3><b>Membuat Pie Chart</b></h3>
<p>
    untuk membuat pie cart, menggunakan fungsi 'plt.pie()', carannya sebagai berikut: 
</p>
<ol>
    <li>memanggil fungsi plt.pie()</li>
    <li>fungsi diisi dengan dataset, label, posisi awal mulai lingkaran, dan jumlah angka dibelakang koma</li> 
    <li>memberikan title chart</li> 
    <li>membuat legenda tentang keterangan pie chart</li> 
</ol>
<h4><b>Hasil</b></h4>
<p>
    dari hasil visualisasi data, didapat bahwa rating '16+' memiliki presentase paling besar yaitu sebesar 30.8%, disusul rating '18+' sebesar 26.6%, dan seterusnya. persentase terkecil didapat oleh rating '13+' yaitu 0.3%
</p>

```python
dc = ds.dropna(subset=['Age'])

pc = dc['Age'].value_counts().rename_axis('Age').reset_index(name='Id_count')
plt.figure(figsize=(5,5))
plt.pie(pc.Id_count, labels=pc.Age, startangle=90, autopct='%.1f%%')
plt.title('Persentase Rating')
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
plt.show()
```
